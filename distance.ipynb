```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
df = pd.read_csv("/Users/adithya/OneDrive/NTU/Y3S2/MDP/Image-Distance.csv")
# df = df[['num_blocks', 'offset (L is positive)', "x", "y", "image_w", "image_h"]]
```

```python
df.describe()
```

```python
df = df.replace(np.nan, 100)
df["dist"] = df["dist"].astype('category')
df["offset"] = df["offset"].astype('category')
df
```

```python
df["aspect_ratio"] = df["w"] / df["h"]
df.describe()
```

```python
# df = df.sample(frac = 1)

X = df[["x", "y", "w", "h"]]

Y = df [["dist", "offset"]].astype("category")


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Y1 = df["num_blocks"]
# Y2 = df["offset (L is positive)"]
len (X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
Y.describe()
```

## Random Forest

```python
clf1 = RandomForestClassifier()
clf2 = RandomForestClassifier()

clf1.fit(X_train, Y_train["dist"])
clf2.fit(X_train, Y_train["offset"])
```

```python
clf1.score(X_train, Y_train["dist"]), clf2.score(X_train, Y_train["offset"])
```
